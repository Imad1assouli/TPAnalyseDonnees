```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data1 = pd.read_csv(r"C:\Users\Ahmed\Desktop\TPAnalyseDonnees\Automobile_data.csv")
```

```python
#Sélectionner les variables explicatives
data=data1[["highway-mpg","engine-size","horsepower","curb-weight"]]
```

```python
#nettoie les données numériques en convertissant les valeurs non numériques en NaN et en remplissant les valeurs manquantes avec la moyenne de leur colonne respective.
for column in data.columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')
    mean = data[column].mean()
    data[column].fillna(mean, inplace=True)
```

```python
print(data)
```

```python
means=data.mean()
means
```

```python
stds=data.std()
stds
```

```python
data=(data-means)/stds
stds
```

```python
print(data)
```

```python
#les composantes principales

pca=PCA()

cp=pca.fit_transform(data)
print(cp)
```

```python
print("Principal Components:")
for i in range(cp.shape[1]):
    print(f"PC{i+1}: {', '.join(str(x) for x in cp[:,i])}")
```

```python
# Dessiner le nuage de points pour les deux premières composantes principales

plt.scatter(cp[:, 0], cp[:, 1])
plt.title("Nuage de points des deux premières composantes principales")
plt.xlabel("Première composante principale")
plt.ylabel("Deuxième composante principale")
plt.show()
```

```python
for x,row in data1.iterrows():
    if row['price']=='?' :
        data1.drop(x,inplace=True)
```

```python

# Ajouter la colonne "price" aux données projetées
cp_with_price = np.hstack((cp, data.iloc[:,-1].values.reshape(-1,1)))
```

```python
# Visualiser les données projetées en utilisant la colonne "price"
sns.scatterplot(data=pd.DataFrame(cp_with_price, columns=["p1","p2","p3","p4","price"]), x="p1", y="p2", hue="price")
plt.title("Nuage de points des deux premières composantes principales avec la colonne 'price'")
plt.xlabel("Première composante principale")
plt.ylabel("Deuxième composante principale")
plt.show()
```

```python
# Matrice de corrélation
corr_matrice = np.corrcoef(data.T)
valc, vecc = np.linalg.eig(corr_matrice)
```

```python
# Calcul et Affichage du cercle de corrélation


fig, ax = plt.subplots(figsize=(10, 10))
sns.set(font_scale=1.5)
sns.set_style('white')
circle = plt.Circle((0, 0), radius=1, color='black', fill=False)
ax.add_artist(circle)

for i in range(len(vecc)):
    ax.arrow(0, 0, vecc[i, 0], vecc[i, 1], head_width=0.05, head_length=0.1, fc='k', ec='k')
    ax.text(vecc[i, 0]*1.2, vecc[i, 1]*1.2, data.columns[i], color='k', ha='center', va='center')

ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_aspect('equal', adjustable='box')
plt.title("Cercle de corrélation")
plt.show()
```

```python
covariance_matrix=np.cov(cp.T)
valp,vecp=np.linalg.eig(covariance_matrix)
```

```python
n = cp.shape[0]
indices = np.random.randint(0, n, size=10)
quality = []
for idx in indices:
    projected_point = cp[idx].dot(vecp[:, :2])
    reconstructed_point = projected_point.dot(vecp[:, :2].T)
    error = np.linalg.norm(cp[idx] - reconstructed_point)
    quality.append(1 - error**2/np.linalg.norm(cp[idx])**2)

print(f"Qualité de projection pour 10 points aléatoires: {quality}")
```
